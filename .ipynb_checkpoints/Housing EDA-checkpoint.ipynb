```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv(r'./train.csv', index_col=0)
test = pd.read_csv(r'./test.csv', index_col=0)
train.shape, test.shape
```

```python
qu_train = train.copy()
qu_train.SalePrice = pd.qcut(train.SalePrice, [0, 0.25, 0.5, 0.75, 1.], labels = [25, 50, 75, 100]).astype('O')
```

```python
x = train['OverallQual']
y = train['SalePrice']
plt.scatter(x, y, marker = 'o', color = 'gold')
```

```python
train.groupby('OverallQual')['SalePrice'].agg(np.mean).sort_values(ascending = False)
```

```python
x = train['OverallCond']
y = train['SalePrice']
plt.scatter(x, y, marker = 'o', color = 'silver')
```

```python
train.groupby('OverallCond')['SalePrice'].agg(np.mean).sort_values(ascending = False)
```

```python
x = train['GrLivArea']
y = train['SalePrice']
plt.scatter(x, y, marker = 'o', color = 'green')
```

```python
qu_train.groupby('SalePrice')['GrLivArea'].agg(np.mean).sort_values(ascending = False)
```

```python
x = qu_train['SalePrice']
y = train['GarageArea']
sns.swarmplot(x, y, data = train)
```

```python
qu_train.groupby('SalePrice')['GarageArea'].agg(np.mean).sort_values(ascending = False)
```

```python
x = qu_train['SalePrice']
y = train['YearBuilt']
sns.swarmplot(x, y, data = train)
```

```python
x = train['LotArea']
y = train['SalePrice']
plt.scatter(x, y, marker = 'o', color = 'brown')
plt.axis([0, 100000, 0, 800000])
```

```python
qu_train.groupby('SalePrice')['LotArea'].agg(np.mean).sort_values(ascending = False)
```

```python
x = train['KitchenQual']
y = train['SalePrice']

sns.swarmplot(x, y, data = train)
```

```python
x = train['MoSold']
plt.hist(x, color = 'black', bins = 30, density = True)
plt.xlabel('Month sold')
plt.ylabel('Frequency')
```
